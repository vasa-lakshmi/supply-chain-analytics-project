feature engineering

```python
# Inventory Metrics
df['Inventory_Turnover'] = (
    df['Monthly_Sales'] /
    df['Current_Stock']
)

df['Stock_Utilization'] = (
    df['Current_Stock'] /
    df['Warehouse_Capacity']
)

# Delivery Metrics
df['Delivery_Efficiency'] = (
    df['Shipping_Time_Days'] +
    df['Order_Processing_Time']
)

# Cost Metrics
df['Total_Cost'] = (
    df['Operational_Cost'] +
    df['Storage_Cost'] +
    df['Transportation_Cost']
)

df['Estimated_Profit'] = (
    df['Monthly_Sales'] -
    df['Total_Cost']
)

# Demand & Risk Metrics
df['Demand_Gap'] = (
    df['Demand_Forecast'] -
    df['Current_Stock']
)

df['Risk_Score'] = (
    df['Stockout_Risk'] +
    df['Return_Rate'] +
    df['Damaged_Goods']
)
```

Exploratory Data Analysis

Sales Trends
Which product categories generate the highest sales?

```python
sales_by_category = df.groupby(
    'Product_Category'
)['Monthly_Sales'].sum().sort_values(
    ascending=False
)

sales_by_category.plot(kind='bar')

plt.title('Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')

plt.show()
```

INSIGHT
 Apparel and Electronics categories contribute the highest monthly sales, indicating strong customer demand in these segments.

SOLUTION:

```python
df['Sales_Efficiency']=(
    df['Monthly_Sales']/
    df['Demand_Forecast']
)
```

```python
sales_eff =df.groupby(
    'Product_Category'
)['Sales_Efficiency'].mean().sort_values(
    ascending=False
)
sales_eff.plot(kind='bar')
```

INSIGHT:
All product categories show strong sales efficiency, indicating that sales are consistently exceeding demand forecasts. Apparel has the highest sales efficiency, making it the strongest-performing category, while Furniture shows slightly lower performance compared to other categories.

```python
RECOMMENDATION:
Since all product categories are performing above forecasted demand, the business should maintain inventory availability across categories. Additional marketing efforts and inventory allocation can be prioritized for Apparel products, while performance monitoring should be conducted for Furniture products to identify potential growth opportunities.
```

DELIVERY PERFORMANCE

INSIGHT
Delivery efficiency across warehouses remains relatively consistent, although a few warehouses show slightly higher delivery times that may require operational monitoring.

SOLUTION:

```python
delivery =df.groupby('Location')['Delivery_Efficiency'].mean()
delivery.plot(kind='bar')
ascending= False
```

INSIGHT:Inventory health varies across locations, with San Francisco and Denver showing the lowest inventory gap values, indicating better alignment between stock levels and demand forecasts. Houston and Dallas show comparatively higher inventory gap values, suggesting a greater mismatch between inventory availability and customer demand.

RECOMMENDATION:Better inventory alignment can reduce storage costs, minimize stock shortages, improve warehouse utilization, and increase overall operational efficiency.
